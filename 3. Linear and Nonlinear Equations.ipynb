**Solve for eigenvalues and eigenvectors of a symmetric matrix**

1. Use numpy's $\tt linalg$ package to find the eigenvalues and eigenvectors of the ff. symmetrix matrix *(10 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
8 & -2 & -20 \\
-2 & -3 & -9 \\
-20 & -9 & -3 
\end{pmatrix} \end{equation} $

2. Check your answer numerically; that is, check that $\bf AV = VD$, where the $N$ x $N$ matrix $\bf V$ has the eigenvectors as its columns and $\bf D$ is the diagonal matrix with the eigenvalues $\lambda_i$ as its diagonal entries. *(5 pts.)*

```python
#Problem 2.1
import numpy as np
A=np.array([[8,-2,-20],[-2,-3,-9],[-20,-9,-3]])
eigen_val,eigen_vec=np.linalg.eigh(A)
print("eigenvalues:")
print(eigen_val)
print("eigenvectors:")
print(eigen_vec)
```

**Problem 2.1 Discussion**\
The eigenvalues and their corresponding eigenvectors were calculated using $\tt np.linalg.eigh$. The column number of the eigenvalue is also the column number of its eigenvector on a separate matrix.

```python
#Problem 2.2
V=eigen_vec
D=np.diag(eigen_val)
print("AV =")
print(np.matmul(A,V))
print("VD =")
print(np.matmul(V,D))
```

**Problem 2.2 Discussion**\
Initially, the eigenvalues were placed diagonally on matrix **D** using the function $\tt np.diag$, and matrix **V** held the eigenvectors. Using the function $\tt np.matmul$ for matrix multiplcation, it was found that **AV** is indeed equal to **VD**.

**Solve a nonlinear equation using the relaxation method**

Consider the equation $x=1- e^{-cx}$, where $c$ is a known parameter and $x$ is unknown. This equation arises in a variety of situations, including the physics of contact processes, mathematical models of epidemics, and the theory of random graphs.

1. Write a program to solve this equation for $x$ using the relaxation method for the case $c = 2$. Calculate your solution to an accuracy of at least $10^{-6}$. *(10 pts.)*

2. Modify your program to calculate the solution for values of $c$ from 0 to 3 in steps of 0.01. *(10 pts.)* 

3. Make a plot of $x$ as a function of $c$. Describe the different regimes that you see. *(5 pts.)*

```python
#Problem 4.1
import numpy as np
import matplotlib.pyplot as plt

accuracy = 10 ** -8
def relax(c):
    f = lambda x: 1 - np.exp(- c * x)
    def error(x, x1):
        return (x - x1) / (1 - 1 / (c * np.exp(-c * x)))
    x = 0.1
    x1 = f(x)
    while(abs(error(x, x1)) > accuracy):
        x, x1 = x1, f(x1)
    return x1
print("x =",relax(2))
```

**Problem 4.1 Discussion**\
I first set the arbitrary initial $x$ value to 0.1 and $c=2$. This will be evaluated to the equation, and the result will become the new $x$ value. To account for the errors of this evaluation, a 'while' loop was used to continuously evaluate over and over the output $x$ value back to the equation until the error is already less than the accuracy, which I set to $10^{-8}$. It was found that $x\approx0.796812125619401$.

```python
#Problem 4.2 & 4.3
c = np.linspace(0, 3, 300)
c = np.delete(c,0)
x = list(map(relax, c))
plt.plot(c, x, 'k.')
plt.xlabel('$c$')
plt.ylabel('$x$')
plt.title("$c$ vs $x$ plot")
plt.show()
```

**Problem 4.2 & 4.3**\
Modifying the previous program in Problem 4.1, the predefined function for the relaxation method was used to map the iterations ranging from $c=0$ to $c=3$ in steps of 0.01. This yielded a graph with a linear regime from $c=0$ to $c=1$, and then it proceeds to an increasing trend which seems to be plateauing on a certain $x$ value.

**Physics Application: Calculating the Lagrange point**

_Refer to Exercise 6.16 of the Newman text._ The $L_1$ Lagrange point is the point between the Earth and the Moon at which a satellite will orbit the Earth in perfect synchrony with the Moon, staying always in between the two. 

Assuming circular orbits, and assuming that the Earth is much more massive than either the Moon or the satellite, the distance $r$ from the center of the Earth to the $L_1$ point satisfies
$$ \begin{equation}
\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2 r,
\end{equation} $$
where $R$ is the distance from the Earth to the Moon, $M$ and $m$ are the Earth and Moon masses, $G$ is Newton's gravitational constant, and $\omega$ is the angular velocity of both the Moon and the satellite.

[EDITED 10/8/2020 - changes in **BOLD**] Use **either** (i) Newton's method **or** (ii) the secant method to solve for the distance r from the Earth to the $L_1$ point. Compute a solution accurate to at least four significant figures. (30 pts.) **BONUS:** *Do the second method for an extra 30 points.*

```python
#Problem 5 - Newton's method
G = 6.674 * 10 ** -11  
M = 5.974 * 10 ** 24 
m = 7.348 * 10 ** 22
R = 3.844 * 10 ** 8  
omega = 2.662 * 10 ** -6 
accuracy = 10 ** -6
f = lambda r: G*M/r**2-G*m/(R-r)**2-omega**2*r
f_der = lambda r:-2*G*M/r**3-2*G*m/(R-r)**3-omega**2
def newton(f, f_der, x_0, accuracy):
    x = x_0
    delta = f(x) / f_der(x)
    while abs(delta) > accuracy:
        delta = f(x) / f_der(x)
        x -= delta
    return x
r = newton(f, f_der, 3 * 10 ** 8, accuracy)
print("r =",r)
```

**Problem 5 Discussion**\
Initially, a function $\tt f$ was defined which is equal to the given relation with all the terms moved to one side of the equation. The derivative of $\tt f$ ($\tt f\_der$) was also predetermined. A function was then defined to carry out Newton's method to find the value of $r$. This method starts with an arbitrary starting value which I chose to be $3\times10^8$. This value will be evaluated to $\tt f$ and $\tt f\_der$. The quotient of these two evaluations will determine the delta parameter which will be subtracted from the starting $r$ value. This new $r$ value will then be continuously evaluated again and again as before using a 'while' loop until delta is less then the accuracy, which I chose to be $10^{-6}$. Finally, the Lagrange point $L_1$ was found to be located at $r\approx326045071.66535544\; \text{m}$ from the Earth.

```python
#Problem 5 - Secant method (bonus)
def f_der(x1,x2):
  return (f(x2)-f(x1))/(x2-x1) #approximation for the derivative
def secant(f, f_der, a,b, accuracy):
    x1 =a
    x2=b
    delta = f(x2) / f_der(x1,x2)
    while abs(delta) > accuracy:
        delta = f(x2) / f_der(x1,x2)
        x1=x2
        x2 -= delta
    return x2
r=secant(f,f_der,3*10**8,3.1*10**8, accuracy)
print("r =",r)
```

**Problem 5 (bonus) Discussion**\
Another method was also used to calculate $r$, namely the secant method. This is an extended version of Newton's method. Instead of using the predefined derivative of $\tt f$, an approximation was made instead using two arbitrary $r$ values. I chose these values to be $3\times10^8$ and $3.1\times10^8$. The process after this will be the same as the rest of Newton's method, and the same accuracy of $10^{-6}$ was also used. It was observed that the resulting $r$ value is exactly the same as the output of Newton's method.
