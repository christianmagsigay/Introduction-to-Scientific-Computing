### PROBLEM 1
**Fibonacci sequence**

1. Write a function that returns the nth Fibonacci number. *(10 pts.)*

2. Print the nth Fibonacci number, for $n = 1, 10, 10^3, 10^6, 10^9$. *(5 pts.)*

````python
#Problem 1.1
#this will supply the succeeding fibbonacci numbers after the first two
def fibo (n):
  numbers = [1,1]
  for x in range (n-2):
    numbers.append (numbers[-1]+numbers[-2]) # this adds the next number to the set, which is just the sum of the previous two
    del numbers[:-3]                         #deletes previous entries (except the last two) to save memory
  return numbers[-1]                         #gives the last number of the set


#Problem 1.2
print(f"1st fibonacci number: {fibo(1)}")
print(f"10th fibonacci number: {fibo(10)}")
print(f"10^3th fibonacci number: {fibo(10**3)}")
print(f"10^6th fibonacci number: {fibo(10**6)}")
````

**PROBLEM 1 DISCUSSION:**

Initially, a function was defined to produce a set of fibonacci numbers. The first 2 numbers were given. The next number was determined by adding the previous two numbers. By using a 'for' loop, new entries were added to the given set. All elements except the last 2 were deleted in each loop to save memory. Lastly, The corresponding fibonacci number for $1, 10, 10^3$ and $10^6$ were determined.

  Unfortunately at $n = 10^9$, this code seems to run in an endless loop. This has something to do with the magnitude of the fibonacci number. $10^9$ runs through the loop 1000 times more than $10^6$, so it needs print out an even larger number  this process takes too long.

### PROBLEM 2
**Curve plotting**

Make a plot of the so-called deltoid curve, which is defined parametrically by the equations $$x = 2 \cos\theta+ \cos2\theta, y = 2 \sin\theta - \sin 2\theta,$$
where $0 \le \theta < 2\pi$. 

1. Take a set of values of $\theta$ between zero and 2$\pi$ and calculate $x$
and $y$ for each from the equations above. *(10 pts.)*

2. Make a plot $y$ as a function of $x$. *(10 pts.)*

````python
import numpy as np
import matplotlib.pyplot as plt
a=0.0
b=2*np.pi

#Problem 2.1
#set of  values:
theta = np.linspace(a,b,100,endpoint=False).tolist() #.tolist() --> in order to print only a single-line set
                                                     #endpoint=False --> removes 2 from the set of values

x,y=[],[]
for i in range(100):
  x.append(2*np.cos(theta[i])+ np.cos(2*theta[i])) #set of x values for each  value 
  y.append(2*np.sin(theta[i])- np.sin(2*theta[i])) #set of y values for each  value  

print(f"\u03B8 values = {theta}")  # values
print(f"x = {x}")                  #x values
print(f"y = {y}")                  #y values


````

**PROBLEM 2.1 DISCUSSION:**

First, 100 values of $\theta$ was generated ranging from $0$ to $2\pi$. Using the given parametric equation, the values of x and y were determined for each of the $\theta$ values. Then, two sets were made using 'for' loops containing all the x and y values corresponding to each $\theta$ value.

````python
####note: run first the previous code before running this one###

#Problem 2.2
#plot of the deltoid curve
plt.plot(x,y)
plt.title('Deltoid curve')
plt.show()
````

**PROBLEM 2.2 DISCUSSION:**

The graph for the values of x and y was generated using the plot function in Matplotlib. The graph verifies that the parametric equation was for a deltoid curve.

### PROBLEM 5

**Calculating the sum of a simple series**

In spite of the power of your trusty computer, calculating the sum of even a simple series may require some thought and care. Consider the two series

$$\begin{equation}
S^{\rm (up)} = \sum_{n=1}^{N} \frac{1}{n},\ S^{\rm (down)} = \sum_{n=N}^{1} \frac{1}{n}.
\end{equation}
$$

Both series are finite as long as $N$ is finite, and when summed analytically both give the same answer. Nonetheless, because of round-off error, the numerical value of $S^{\rm (up)}$ will not be precisely that of $S^{\rm (down)}$.

1. Calculate $S^{\rm (up)}$ and $S^{\rm (down)}$ as functions of $N$. *(10 pts.)*

2. Make a log-log plot of $(S^{\rm (up)}-S^{\rm (down)})/(|S^{\rm (up)}|+|S^{\rm (down)}|)$ vs. $N$. *(10 pts.)*

3. Observe the linear regime on your graph and explain why the downward sum is generally more precise. *(5 pts.)*

````python
import matplotlib.pyplot as plt
import numpy as np

#Problem 5.1
#evaluates the series S^(up)
def series_up (n):                 #This starts adding from n=1 to n=N
  S_up=[0]                       #set that will contain the partial sums of the series
  for n in range(n):
    S_up.append(S_up[-1]+1/(n+1))  #adds the next entry to the set
  return S_up[-1]                 #returns the final sum

#evaluates the series S^(down)
def series_down (k):               #This starts adding from n=N to n=1
  S_down = [0]                    #set that will contain the partial sums of the series
  while k != 0:
    S_down.append(S_down[-1]+1/k)  #adds the next entry to the set
    k -= 1
  return S_down[-1]               #returns the final sum

#N input
N=np.random.randint(1,10000)

#increases decimal places
a="{0:.100f}".format(series_up(N))
b="{0:.100f}".format(series_down(N))

print(f"N = {N}")
print(f"S^(up) = {a}")
print(f"S^(down) = {b}")
````

**PROBLEM 5.1 DISCUSSION:**

A 'for' loop was used to evaluate the series $S^{\rm(up)}$ for a random value of $N$. A 'while' loop on the other hand, was used to evaluate the series $S^{\rm(down)}$. It was observed that the sum of these two equations are slightly different at certain values of $N$ especially at larger values.

````python
####note: run first the previous code before running this one###

#Problem 5.2

N=5000                            # number of N

#evaluation
up,down=[],[]                      #sets that will contain the evaluation from 1 to N
for i in range (N):
  up.append(series_up(i+1))        #adds entry to the set 'up'
                                   #generates the evaluation of S^(up) from 1 to N
  down.append(series_down(i+1))    #adds entry to the set 'down'
                                   #generates the evaluation of S^(down) from 1 to N

#conversion to array
up=np.array(up)
down=np.array(down)


y=(up-down)/(abs(up)+abs(down)) #Solves the given equation
x=np.arange(1,N+1)              #generating a set from 1 to N

#log-scaled plot
fig = plt.figure()
plt.plot(x,y, '.',c="firebrick")
plt.yscale('log')
plt.xscale('log')
plt.xlabel("N")
plt.ylabel('$(S^{(up)}-S^{(down)})/(|S^{(up)}|+|S^{(down)}|)$')
plt.show()
````

**PROBLEM 5.2 & 5.3 DISCUSSION:**

Initially, $N$ was given a value of $5000$. Using the function defined previously, a set was made to contain the sums of $N=1$ until $N=5000$ evaluated to $S^{\rm(up)}$. The same goes for $S^{\rm(down)}$. These were then used to find the value of the relative difference for each value of $N$. Then, the results from the given quantity and the corresponding $N$ values were plotted on a log-scaled graph. 

Visually, the individual markers resemble a series of diagonal lines piled on top of each other.

$S^{\rm(down)}$ evaluates from $n=N$ to $n=1$, so it adds up terms from least to greatest in value. This provides a more precise evaluation, since it starts with values that has a lesser precision and it continuously adds up values with higher precision. If you start adding small values first, it could sum up to a significantly large value that can affect the final sum. Adding the largest value first would lose precision, especially if the there's a large difference between the succeeding term, since the smaller values might not be significant enough to affect the total sum due to rounding errors.


````python
#Additional

y=(up-down)/(abs(up)+abs(down))

#Solving the absolute value of the equation
z=abs((up-down)/(abs(up)+abs(down)))

#generating a set from 1 to N
x=np.arange(1,N+1)

#log-scaled plot
plt.plot(x,z,'.', c='seagreen')
plt.plot(x,y, '.',c="firebrick") 

plt.yscale('log')
plt.xscale('log')
plt.xlabel("N")
plt.ylabel('$(|S^{(up)}-S^{(down)}|)/(|S^{(up)}|+|S^{(down)}|)$')

plt.show()
````

**ADDITIONAL DISCUSSION:**

The given quantity for the relative difference also generates negative values. These values can't be plotted on a log-scaled graph, so the absolute value of the quantity was determined to account for these missing values. This new quantity was represented in green markers and the original quantitiy was superimposed on it. The visible green markers represents the contribution of the origially negative values. It was observed that these values (up to a certain degree) also follow the series of diagonal lines made by the original quantity.
